```python
import pandas as pd
import klib

from src.data_loader import DataLoader
```

```python
data_loader = DataLoader(path="../data/raw/restaurant_dataset.csv")
```

```python
df = data_loader.load()
```

# Overview

```python
print(df.shape)
print(df.head())
print(df.info())
print(df.describe())
```

# Missing values

```python
import seaborn as sns
import matplotlib.pyplot as plt

print(df.isnull().sum())
sns.heatmap(df.isnull(), cbar=False, cmap="Reds")
plt.title("Bo'sh qiymatlar")
plt.show()
```

# Correlation

```python
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm")
plt.title("Correlation matrix")
plt.show()
```

# Distribution

```python
import matplotlib.pyplot as plt

df["aggregate_rating"].hist(bins=20)

plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Aggregate Rating Distribution")

plt.show()
```
