# Baseline Model
LinearRegression and RandomForest — `aggregate_rating` prediction

```python
import pandas as pd
from src.trainer import Trainer
```

## 1. Data Load

```python
x_train = pd.read_csv('../data/preprocessed/baseline/x_train.csv')
x_test  = pd.read_csv('../data/preprocessed/baseline/x_test.csv')
y_train = pd.read_csv('../data/preprocessed/baseline/y_train.csv').squeeze()
y_test  = pd.read_csv('../data/preprocessed/baseline/y_test.csv').squeeze()

print(f'x_train: {x_train.shape}, x_test: {x_test.shape}')
```

## 2. Linear Regression

```python
trainer = Trainer(version='baseline')

trainer.train(x_train, y_train, model_name='linear_regression')

metrics_lr = trainer.evaluate(x_test, y_test)
print(metrics_lr)

trainer.save_model()
trainer.save_results(metrics_lr)
```

## 3. Random Forest

```python
trainer = Trainer(version='baseline')

trainer.train(x_train, y_train, model_name='random_forest')

metrics_rf = trainer.evaluate(x_test, y_test)
print(metrics_rf)

trainer.save_model()
trainer.save_results(metrics_rf)
```
